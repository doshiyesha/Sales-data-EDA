```python
merchant_sales = df[["merchant_name","merchant_rating","Total amt Eur"]]
merchant_sales.head(20)
merchant_sales.sort_values("merchant_rating", ascending=False)
```

```python
merchant_sales.describe()
```

```python
filter1 = (merchant_sales["merchant_rating"] > 3.8) & (merchant_sales["merchant_rating"] <= 4.0)
filter2 = (merchant_sales["merchant_rating"] > 4.0) & (merchant_sales["merchant_rating"] <=4.057)
filter3 = (merchant_sales["merchant_rating"] > 4.057) & (merchant_sales["merchant_rating"] <=4.093)
filter4 = (merchant_sales["merchant_rating"] > 4.093) & (merchant_sales["merchant_rating"] <=4.155)
filter5 = (merchant_sales["merchant_rating"] > 4.155) & (merchant_sales["merchant_rating"] <=4.3)

Lowest_rating = merchant_sales[filter1]
Low_rating = merchant_sales[filter2]
average_rating = merchant_sales[filter3]
above_average = merchant_sales[filter4]
good_rating = merchant_sales[filter5]

Lowest_rating["Rating Category"] = "Lowest Rating"
Low_rating["Rating Category"] = "Low Rating"
average_rating["Rating Category"] = "Average Rating"
above_average["Rating Category"] = "Above Average"
good_rating["Rating Category"] = "Good Rating"

list = [Lowest_rating,Low_rating,average_rating,above_average,good_rating]
```

```python
for i in list:
    print(i.shape)
```

```python
df2 = pd.concat(list)
df2.head()
```

```python
Total_sales_byRating = df2.groupby("Rating Category")["Total amt Eur"].agg(['sum','mean'])
Total_sales_byRating
Total_sales_byRating.plot.bar()
```

```python
merchant_sales["Merchant Rating Category"] = pd.cut(merchant_sales["merchant_rating"],bins=5,labels=("Lowest rating","Low rating","Average rating","Above-average","Good"))
merchant_sales.head(20)
```

```python
merchant_sales2 = merchant_sales.loc[merchant_sales["Total amt Eur"]>=50000]
merchant_sales2 = merchant_sales2.sort_values("Total amt Eur",ascending=False)


from matplotlib import pyplot as plt
a4_dims = (12.7, 9.27)
fig, ax = plt.subplots(figsize=a4_dims)
import seaborn as sns
sns.barplot(x = "Total amt Eur",
            y = "merchant_name",
            data = merchant_sales2)
plt.title("Sales by Merchant")
plt.show()
```

```python
Merchant_totalsales_byRating_category = merchant_sales.groupby("Merchant Rating Category")[["merchant_rating","Total amt Eur"]].agg(['sum','mean'])
Merchant_totalsales_byRating_category
```
